```python
# Replace the null values of country column with mode value
df['country'].fillna(df.country.mode().to_string(), inplace=True)

# Replace the null values of children column with mean value
df['children'].fillna(round(df.children.mean()), inplace=True)
```

```python
df.head()
```

```python
# Let's now check for null or missing value in our dataframe
df.isnull().sum()
```

```python
# Columns that is required too keep in another dataframe for future analysis
columns_to_keep = ['hotel', 'is_canceled', 'lead_time', 'arrival_date_year', 'arrival_date_month', 'country', 'adults', 'children', 'babies', 'stays_in_weekend_nights', 'market_segment', 'total_of_special_requests']
df = df[columns_to_keep]
# Let's display the information of the dataframe
df.info()
```

```python
# Let's see in what is booking distribution year wise
year_wise_dis = (df['arrival_date_year'].value_counts())
year_wise_dis
```

```python
# Let's see above data graphically using count plot
plt.figure(figsize=(7,5))
sns.countplot(data=df, x='arrival_date_year')
plt.title("Booking Distribution of Hotels Year Wise ", fontsize = 22)
plt.ylabel('Number Of Bookings',fontsize = 15)
plt.xlabel('Year Of Hotels Booking',fontsize = 15)
plt.show()
```

```python
# Now let see booking ratio of hotels year wise
year_wise_booking_ratio = df["arrival_date_year"].value_counts()
year_wise_booking_ratio.plot.pie(figsize = (12,6),fontsize = 16, autopct="%.1f%%")
plt.title('Year wise Booking Ratio of Hotels in percentage',fontsize = 22)
plt.ylabel('Bookings Percentage',fontsize = 15)
plt.xlabel('Year Of Hotels Booking',fontsize = 15)
pass
```

```python
 #First, we use lineplot to to display the monthly trend for both City and Resort Hotel simultaneously.

total_bookings = df["arrival_date_month"].value_counts()
total_bookings.plot.pie(figsize = (14,6),fontsize = 12, autopct="%.1f%%")
plt.title('Month Wise Booking Distribution For Both  City Hotel and Resort Hotel Simultaneously ',fontsize = 24)
plt.ylabel('Bookings In %',fontsize = 24)
plt.xlabel('Month Of Bookings',fontsize = 22)
pass
```

```python
# So we need to split the data in 2 sets for City And Resort Hotel respectively so we can easily understand trends for City and Resort Hotel seperately.

city_hotel = df[df['hotel']=='City Hotel']
resort_hotel = df[df['hotel']=='Resort Hotel']
```

```python
# Just because Bookings in 2016 Was Higher then other years so to check the distribution of booking on monthly basis 
# we will take only the year 2016 in consideration for both hotel and resort 

city_hotel_in_2016 = city_hotel[city_hotel['arrival_date_year']==2016]
resort_hotel_in_2016 = resort_hotel[resort_hotel['arrival_date_year']==2016]

fig, axes = plt.subplots(1, 2, figsize=(22, 6))
fig.suptitle('Distribution Of Monthwise Bookings  for City and Resort Hotel in 2016', fontsize=20)

# Monthwise Bookings Distribution for Resort Hotel in 2016
sns.countplot(ax = axes[0], data = city_hotel_in_2016, x = "arrival_date_month")
axes[0].set_title('Monthwise Bookings Distribution for City Hotel in 2016', fontsize=15)

# Monthwise Bookings Distribution for Resort Hotel in 2016
sns.countplot(ax = axes[1], data=resort_hotel_in_2016, x= "arrival_date_month" )
axes[1].set_title('Monthwise Bookings Distribution for Resort Hotel in 2016', fontsize=15)
plt.show()
```

```python
# Let's see the booking difference between city and resort hotel
booking_diff_of_hotels = df['hotel'].value_counts()
booking_diff_of_hotels
```

```python
# Let's compare booking difference between city and resort hotel on plot
plt.figure(figsize=(8,5))
sns.countplot(data=df, x = 'hotel')
plt.title("Booking Comparison Between City And Resort Hotel", fontsize = 22)
plt.ylabel('Number Of Bookings',fontsize = 15)
plt.xlabel('Year Of Hotels Booking',fontsize = 15)
plt.show()
```

```python
# Let's see how many bookings is cancelled and not cancelled in percent(%)
plt.figure(figsize=(12,6))
status = [df.is_canceled.sum(),len(df)-df.is_canceled.sum()]
labels=['Canceled Booking In %','Not Canceled Bookings In %']
plt.pie(status, labels=labels, autopct='%.0f%%')
plt.title('Comparison of  Cancelled and Not Cancelled Bookings In Percent (%)', fontsize=22)
plt.show()
```

```python
# cancelled and not cancelled booking distributed for City Hotel And Resort Hotel
plt.figure(figsize=(9,6))
sns.countplot(data=df, x='hotel', hue='is_canceled')
plt.title("Cancelled and Not Cancelled Bookings Distributed For Both City And Resort Hotel", fontsize = 22)
plt.xlabel("Hotel", fontsize = 15)
plt.ylabel("Number Of Bookings", fontsize = 15)
plt.show()
```

```python
# Let's see from which top 7 countries mostly guests comes or bookings done

fig, axes = plt.subplots(1, 2, figsize=(22, 7))
fig.suptitle('Distribution Of Countries on the basis Of mostly(maximum) guests comes in both hotels(City And Resort)', fontsize=20)
sns.countplot(ax = axes[0], data = city_hotel_in_2016, x = "country", order = city_hotel_in_2016.country.value_counts().iloc[:7].index)
axes[0].set_title('Distribution Of Countries on the basis of maximum guests comes in City Hotel', fontsize=15)
sns.countplot(ax = axes[1], data=resort_hotel_in_2016, x= "country", order = resort_hotel_in_2016.country.value_counts().iloc[:7].index )
axes[1].set_title('Distribution Of Countries on the basis of maximum guests comes in Resort Hotel', fontsize=15)
plt.show()
```
